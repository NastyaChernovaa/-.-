## Функция одной переменной

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.Symbol('x')
f = 3*x**2   
```

```python
f_lambda = sp.lambdify(x, f, 'numpy')
# Генерация значений x для построения графика
x_vals = np.linspace(-4, 4, 100)
y_vals = f_lambda(x_vals)
plt.plot(x_vals, y_vals, label=str(f))
plt.legend()
```

```python
f_prime = sp.diff(f, x)
x0 = 5
la = 0.01
```

```python
res = []
```

```python
x1=0
while (np.abs(x0) ** 2 + np.abs(x1) ** 2) ** (1/2) > 0.00000001:
    res.append([x0, f.subs(x, x0)])
    x1 = x0 - la * f_prime.subs(x, x0)
    x0 = x1
```

```python
res
```

```python
x = []
y = []
for i in range(len(res)):
    x.append(res[i][0])
    y.append(res[i][1])
```

```python
plt.plot(x, y)
plt.plot(x_vals, y_vals, label=str(f))
plt.scatter(x, y)
```

## Функция двух переменных

```python
x, y = sp.symbols('x y')
f = 3*x**2 + 5*y**2
```

```python
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
f_lambda = sp.lambdify((x, y), f, 'numpy')
Z = f_lambda(X, Y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

plt.show()
```

```python
f = lambda x, y: 3*x**2 + 5*y**2
f_dx = lambda x, y: 6*x
f_dy = lambda x, y: 10*y
```

```python
x0 = 5
y0 = 10
la = 0.1
res = []
```

```python
while True:
    x1 = x0 - la * f_dx(x0, y0)
    y1 = y0 - la * f_dy(x0, y0)
    res.append([x0, y0, f(x0, y0)])
    if np.linalg.norm(np.array([x0 - x1, y0 - y1])) <=  0.00000001:
        break
    x0 = x1
    y0 = y1
```
